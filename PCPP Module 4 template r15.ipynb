#### 3.2 GENERATORS
#### Generator Functions

```python
def mygen():
    print("Inside mygen... Creating new generator")
    yield 'a'
    yield 'b'
    yield 'c'
    

gen1 = mygen()
print(gen1)


# getting values using next()

print(next(gen1))
print(next(gen1))
print(next(gen1))
print(next(gen1))
```

```python
# using generator with a for loop

gen2 = mygen()

for y in gen2:
    print('first loop:', y)

print('end of first loop')


# exhaustible - not reusable
    
for z in gen2:
    print('second loop:', z)

print('end of second loop')
```

```python
# Another example

def mygen2():
    n = 1
    print('First value')
    yield n

    n += 1
    print('Second value')
    yield n

    n += 1
    print('Last value')
    yield n
    

gen3 = mygen2()

# print(next(gen3))    # try running code with/without this line

print("--- for loop ---")
for x in gen3:
    print(x)
```

```python
# Repeat thrice

def repeat_three_times(value):
    yield value
    yield value
    yield value
    
three = repeat_three_times(10)
next(three)
```

```python
# Exercise: repeat() with parameters 
```

```python
# Example: Represent an infinite stream of data, since it cannot be stored in memory 

def all_even():
    n = 0
    while True:
        yield n
        n += 2
        
infinite_evens = all_even()

print(next(infinite_evens))
print(next(infinite_evens))
print(next(infinite_evens))
```

#### Generator expressions

```python
# recall list comprehension

squares_list = [x*x for x in range(1,10)]
print(squares_list)
```

```python
# generator expression

squares = (x*x for x in range(1,10))
print(squares)

print(next(squares))

for square in squares:
    print(square)
```

```python
# rewriting the repeat() generator function

repeater = ("hello" for i in range(4))

print(next(repeater))
```

```python
# processing a large dataset

def has_88(number):    
    return "88" in str(number)

large_data = range(2, 10000000)
eighty_eights = (i   for i in large_data   if has_88(i))


for x in eighty_eights:
    print(x)
```

#### Pattern matching

```python
# String methods
```

```python
# [Extra] glob module
import glob

# returns a list
txt_files = glob.glob("*.txt")
print(txt_files)

# returns a generator
txt_files2 = glob.iglob("*.txt")
print(txt_files2)

print("File names starting with 'new':")
for file in glob.glob('new*.*'):
    print("\t" + file)

print("File names containing digits:")
for file in glob.iglob("*[0-9]*.txt"):
    print("\t" + file)
```
